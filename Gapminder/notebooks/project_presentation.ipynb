```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import imageio 
import os
```

```python
df= pd.read_csv("../data/gapminder_total.csv")
```

```python
df.describe()
```

```python
df_s= df[df["country"] == "India"]
```

```python
df_s
```

```python
df_s["population"].astype(int)
```

```python
df_s.assign(population_b = df_s["population"]/10000000)
df_s
```

```python
x_s= df_s["year"]
a_s= df_s["life expectancy"]
b_s= df_s["fertility"]
c_s=df_s["population_b"]

for i in range(1950,2016):
    df_subset = df_s[df_s["year"] == i]
    x= df_subset["year"]
    a= df_subset["life expectancy"]
    b= df_subset["fertility"]
    c= df_subset["population_b"]
    plt.figure(figsize=(20, 10))
    plt.title("Changes over years")
    plt.grid()
    plt.axis([1945, 2020, 0, 140])
    plt.plot(x, a, color='r', label='life expectancy', marker = "s")
    plt.plot(x, b, color='g', label='fertility', marker = "^")
    plt.plot(x, c, color='b', label='population', marker = "o")
    plt.xlabel("Year")
    plt.ylabel("India")
    plt.legend()
    
    plt.plot(x_s, a_s, color='r')
    plt.plot(x_s, b_s, color='g')
    plt.plot(x_s, c_s, color='b')    
    plt.savefig(f"../data/test/plot_p_{i}.png")
    plt.close()
    
```

```python
images = []

for year in range(1960, 2016):
    filename = f'../data/test/plot_p_{year}.png'
    images.append(imageio.imread(filename))

imageio.mimsave("../data/test.gif", images, fps=10)
```
